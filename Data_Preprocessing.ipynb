```python
missing_values_list = missing_values.to_dict()
missing_values_list
```

```python
# remove columns from df with null values = 19233 from the missing_values_list

columns_to_remove = [key for key, value in missing_values_list.items() if value == 19233]
df_cleaned = df.drop(columns=columns_to_remove)
df_cleaned.head()
```

```python
df_cleaned.columns
```

```python
# remove columns with null values = 19233 from the missing_values_list

cleaned_missing_values_list = {key: value for key, value in missing_values_list.items() if value != 19233}
print(cleaned_missing_values_list)
```

```python
for c in cleaned_missing_values_list:
  if cleaned_missing_values_list[c] != 0:
    df_cleaned[c].interpolate(method="linear", inplace=True)
```
